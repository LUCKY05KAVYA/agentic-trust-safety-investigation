```python
def cross_modal_verification(state):
    text = state["listing"]["title"] + " " + state["listing"]["description"]
    images = state["listing"]["image_paths"]

    scores = []

    for img in images:
        score = clip_score(img, text)
        scores.append(score)

    state["clip_score"] = sum(scores) / len(scores) if scores else 0.0

    return state
```

```python
print("######## NEW UNDERSTANDING NODE ########")

def understanding_node(state: InvestigationState) -> InvestigationState:
    listing = state.listing

    prompt = f"""
Analyze product:

Title: {listing.get('title', '')}
Description: {listing.get('description', '')}
Price: {listing.get('price', '')}
Brand: {listing.get('brand', '')}

Return ONLY valid JSON in this format:

{{
    "violations": [],
    "risk_analysis": "",
    "confidence": 0.6
}}
"""

    messages = [
        SystemMessage(content=SYSTEM_PROMPT),
        HumanMessage(content=prompt)
    ]

    response = llm.invoke(messages)

    print("RAW:", response.content)

    import json
    import re

    raw = response.content

    match = re.search(r"\{.*\}", raw, re.DOTALL)

    if match:
        try:
            parsed = json.loads(match.group())
        except Exception:
            parsed = {
                "violations": [],
                "risk_analysis": "",
                "confidence": 0.6
            }
    else:
        parsed = {
            "violations": [],
            "risk_analysis": "",
            "confidence": 0.6
        }

    raw_violations = parsed.get("violations", [])

    clean_violations = []

    for v in raw_violations:
        if isinstance(v, dict):
            clean_violations.append(
                Violation(
                    type=v.get("type", "unknown"),
                    description=v.get("description", "")
                )
            )
        else:
            clean_violations.append(
                Violation(
                    type="unknown",
                    description=str(v)
                )
            )

    state.violations = clean_violations
    state.risk_analysis = parsed.get("risk_analysis", "")
    state.confidence = parsed.get("confidence", 0.6)

    print("AFTER CLEANING:")
    print(state.violations)
    print([type(v) for v in state.violations])

    return state
```

```python
import numpy as np

def compute_metrics(results):

    total = len(results)

    correct = sum(r["correct"] for r in results)

    decision_accuracy = correct / total

    avg_latency = np.mean([r["latency"] for r in results])

    # confusion matrix
    tp = sum((r["predicted_risk"] == "High" and r["actual_risk"] == "High") for r in results)
    tn = sum((r["predicted_risk"] == "Low" and r["actual_risk"] == "Low") for r in results)
    fp = sum((r["predicted_risk"] == "High" and r["actual_risk"] == "Low") for r in results)
    fn = sum((r["predicted_risk"] == "Low" and r["actual_risk"] == "High") for r in results)

    precision = tp / (tp + fp + 1e-8)
    recall = tp / (tp + fn + 1e-8)
    f1 = 2 * precision * recall / (precision + recall + 1e-8)

    violation_detection = recall  # same as before definition

    manual_reduction = sum(r["predicted_risk"] == "Low" for r in results) / total

    # CLIP metrics (NEW for Ablation 3)
    avg_clip_score = np.mean([r.get("clip_score", 0) for r in results])

    return {
        "Total Samples": total,
        "Decision Accuracy": round(decision_accuracy * 100, 2),
        "Violation Detection Rate": round(violation_detection * 100, 2),
        "Average Latency": round(avg_latency, 2),
        "Manual Review Reduction": round(manual_reduction * 100, 2),
        "Precision (High-risk)": round(precision * 100, 2),
        "Recall (High-risk)": round(recall * 100, 2),
        "F1 Score": round(f1 * 100, 2),
        "Avg CLIP Score": round(avg_clip_score, 4)
    }
```

```python
def policy_risk_node(state: InvestigationState):

    listing = state.listing

    prompt = f"""
    Perform a detailed Trust & Safety analysis on this product listing.

    Title: {listing.get('title')}
    Description: {listing.get('description')}
    Price: ₹{listing.get('price')}
    Brand: {listing.get('brand')}

    Previous findings:
    Inconsistencies: {state.inconsistencies}
    Query Intent: {state.query_intent}

    Return ONLY valid JSON.
    """

    messages = [
        SystemMessage(content=SYSTEM_PROMPT),
        HumanMessage(content=prompt)
    ]

    response = llm.invoke(messages)

    import json, re

    match = re.search(r"\{.*\}", response.content, re.DOTALL)

    parsed = json.loads(match.group()) if match else {}

    raw_violations = parsed.get("violations", [])

    clean_violations = []

    for v in raw_violations:
        if isinstance(v, dict):
            clean_violations.append(
                f"{v.get('type','Unknown')}: {v.get('description','')}"
            )
        else:
            clean_violations.append(str(v))

    state.violations = clean_violations
    state.risk_analysis = parsed.get("risk_analysis", "")
    state.confidence = parsed.get("confidence", 0.7)

    if (
        "leather" in listing.get("description", "").lower()
        and state.inconsistencies
    ):
        state.violations.append("Misleading Material Claims")

    if (
        "counterfeit" in response.content.lower()
        or "fake" in response.content.lower()
    ):
        state.violations.append("Possible Counterfeit")

    state.violations = list(set(state.violations))

    return state
```

```python
def reflection_node(state):
    """Ablation 2: No self-reflection"""

    state.reflection = ""
    state.confidence = 70.0

    return state
```

```python
def decision_node(state: InvestigationState) -> InvestigationState:
    """Node 5: Make final structured decision"""

    print("\n========== DEBUG ==========")
    print("Title:", state.listing.get("title"))
    print("Violations:", state.violations)
    print("Risk Analysis:", state.risk_analysis)
    print("Reflection:", state.reflection)
    print("Confidence:", state.confidence)
    print("===========================\n")

    prompt = f"""
    Based on all analysis, give final Trust & Safety decision.

    Risk Analysis: {state.risk_analysis}
    Violations: {state.violations}
    Reflection: {state.reflection}
    Confidence: {state.confidence}

    Return in clear structured format.
    """

    messages = [
        SystemMessage(content=SYSTEM_PROMPT),
        HumanMessage(content=prompt)
    ]

    response = llm.invoke(messages)

    state.final_decision = {
        "risk_level": "High" if state.violations else "Low",
        "action": "Flag for Manual Review" if state.violations else "Approve",
        "explanation": response.content.strip(),
        "violations": state.violations,
        "confidence": state.confidence
    }

    return state

print("✅ Node 5 (Final Decision) created")
```

```python
def run_investigation(listing_data: Dict):
    """Run the full agent on a product listing"""
    print("🚀 Starting Trust & Safety Investigation...\n")

    initial_state = InvestigationState(listing=listing_data)
    final_state = agent.invoke(initial_state)

    # Handle both Pydantic object and dict return types
    if isinstance(final_state, dict):
        state_dict = final_state
    else:
        state_dict = final_state.dict() if hasattr(final_state, 'dict') else final_state.__dict__

    print("=== FINAL DECISION ===\n")
    decision = state_dict.get('final_decision', {})

    print(f"Risk Level        : {decision.get('risk_level', 'N/A')}")
    print(f"Recommended Action: {decision.get('action', 'N/A')}")
    print(f"Confidence        : {decision.get('confidence', 'N/A')}%")
    print(f"Violations        : {decision.get('violations', [])}")
    print("\nExplanation:")
    print(decision.get('explanation', state_dict.get('risk_analysis', 'No explanation available')))

    # Optional: Print more details
    print("\n--- Additional Info ---")
    print(f"Inconsistencies: {state_dict.get('inconsistencies', [])}")

    return final_state

print("✅ run_investigation function updated")
```

```python
def decision_node(state: InvestigationState) -> InvestigationState:
    """Reliable Structured Decision Node"""

    prompt = f"""
    You are a senior Trust & Safety investigator. Return a clean structured decision.

    Listing:
    Title: {state.listing.get('title')}
    Description: {state.listing.get('description')}
    Price: ₹{state.listing.get('price', 'N/A')}

    Analysis Summary:
    Inconsistencies: {state.inconsistencies}
    Violations so far: {state.violations}
    Risk Notes: {state.risk_analysis[-300:] if state.risk_analysis else 'None'}

    Provide the decision in this exact JSON format:
    {{
      "risk_level": "High/Medium/Low",
      "action": "Approve / Flag for Manual Review / Remove Listing",
      "violations": ["violation1", "violation2"],
      "explanation": "clear explanation here",
      "confidence": 85,
      "citations": ["citation 1", "citation 2"]
    }}
    """

    messages = [
        SystemMessage(content=SYSTEM_PROMPT),
        HumanMessage(content=prompt)
    ]

    response = llm.invoke(messages)
    content = response.content.strip()

    # Try to parse JSON
    import json
    try:
        # Extract JSON if LLM added extra text
        if '{' in content and '}' in content:
            json_str = content[content.find('{'):content.rfind('}')+1]
            decision_dict = json.loads(json_str)
        else:
            decision_dict = json.loads(content)
    except:
        # Fallback
        decision_dict = {
            "risk_level": "High" if state.violations else "Medium",
            "action": "Flag for Manual Review",
            "violations": state.violations or ["Potential Policy Violation"],
            "explanation": content[:800],
            "confidence": 75,
            "citations": []
        }

    state.final_decision = decision_dict
    state.confidence = decision_dict.get("confidence", 75.0)

    return state

print("✅ Decision Node updated with reliable JSON structured output")
```

```python
import time
from collections import defaultdict

def evaluate_agent(dataset, num_samples=None):
    if num_samples is None:
        num_samples = len(dataset)

    results = []
    total_time = 0.0

    correct_risk = 0
    correct_action = 0
    true_positives = 0
    false_positives = 0
    false_negatives = 0
    hallucinated_violations = 0

    print(f"🚀 Running Full Evaluation on {num_samples} samples...\n")

    for item in dataset[:num_samples]:
        start = time.time()

        listing = {
            "title": item["title"],
            "description": item["description"],
            "price": item["price"],
            "brand": item["brand"],
            "image_paths": item.get("image_paths", [])
        }

        final_state = agent.invoke(InvestigationState(listing=listing))
        state_dict = final_state if isinstance(final_state, dict) else getattr(final_state, '__dict__', lambda: final_state)()
        decision = state_dict.get("final_decision", {})

        latency = time.time() - start
        total_time += latency

        pred_risk = decision.get("risk_level", "Medium")
        gt_risk = item["ground_truth"]["risk_level"]

        pred_violations = set(decision.get("violations", []))
        gt_violations = set(item["ground_truth"].get("violations", []))

        # Core Metrics
        if pred_risk == gt_risk:
            correct_risk += 1
        if decision.get("action") == item["ground_truth"].get("action"):
            correct_action += 1

        # Violation Metrics
        if gt_violations and pred_violations:
            true_positives += 1
        elif pred_violations and not gt_violations:
            false_positives += 1
        elif gt_violations and not pred_violations:
            false_negatives += 1

        # Rough Hallucination Check
        if len(pred_violations) > len(gt_violations) + 1:
            hallucinated_violations += 1

        results.append({
            "id": item["id"],
            "predicted_risk": pred_risk,
            "actual_risk": gt_risk,
            "latency": round(latency, 2)
        })

    # Final Metrics
    n = num_samples
    risk_accuracy = (correct_risk / n) * 100
    action_accuracy = (correct_action / n) * 100
    avg_latency = total_time / n

    precision = true_positives / (true_positives + false_positives) * 100 if (true_positives + false_positives) > 0 else 0
    recall = true_positives / (true_positives + false_negatives) * 100 if (true_positives + false_negatives) > 0 else 0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

    print("=== PROFESSIONAL EVALUATION RESULTS ===")
    print(f"Risk Level Accuracy      : {risk_accuracy:.1f}%")
    print(f"Action Accuracy          : {action_accuracy:.1f}%")
    print(f"Violation Precision      : {precision:.1f}%")
    print(f"Violation Recall         : {recall:.1f}%")
    print(f"Violation F1 Score       : {f1:.1f}%")
    print(f"False Negative Rate      : {(false_negatives / n * 100):.1f}%")
    print(f"Hallucination Rate       : {(hallucinated_violations / n * 100):.1f}%")
    print(f"Average Latency          : {avg_latency:.2f} seconds")
    print(f"Simulated Manual Review Reduction : ~{int(risk_accuracy * 0.65)}%")

    return results

print("✅ Enhanced Evaluation with Precision, Recall, F1, False Negatives, Hallucination Rate ready!")
```
